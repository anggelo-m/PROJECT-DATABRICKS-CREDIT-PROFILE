```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("catalogo", "credit_catalog")
dbutils.widgets.text("esquema_source", "bronze")
dbutils.widgets.text("esquema_sink", "silver")
```

```python
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
df_application_record = spark.table(f"{catalogo}.{esquema_source}.application_record")
```

```python
df_application_record = df_application_record.dropna(how="all")\
                        .filter((col("ID").isNotNull()) | (col("AMT_INCOME_TOTAL")).isNotNull())
```

```python

df_application_record_clean = df_application_record.withColumn(
    "age", round(abs(col("DAYS_BIRTH"))/365,0)
).withColumn(
    "employment_years", round(abs(col("DAYS_EMPLOYED"))/365,0)
).drop("DAYS_BIRTH","DAYS_EMPLOYED")

```

```python
df_application_record_clean.write.mode("overwrite").saveAsTable(f"{catalogo}.{esquema_sink}.application_record_transformed")
```
