```sql
%python
dbutils.widgets.removeAll()
```

```sql
%python
dbutils.widgets.text("nameContainer","raw")
dbutils.widgets.text("nameStorage","adlssmartdata1202")
dbutils.widgets.text("catalogo","credit_catalog")
```

```sql
%python

nameContainer = dbutils.widgets.get("nameContainer")
nameStorage = dbutils.widgets.get("nameStorage")
catalogo = dbutils.widgets.get("catalogo")

ruta = f"abfss://{nameContainer}@{nameStorage}.dfs.core.windows.net"
```

### Eliminacion tablas Bronze

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.bronze.application_record;
DROP TABLE IF EXISTS ${catalogo}.bronze.credit_record;
```

```sql
%python
## REMOVE DATA (Bronze)
dbutils.fs.rm(f"{ruta}/tablas/application_record", True)
dbutils.fs.rm(f"{ruta}/tablas/credit_record", True)
```

### Eliminacion tablas Silver

```sql
-- DROP TABLES
DROP TABLE IF EXISTS ${catalogo}.silver.application_record_transformed;
DROP TABLE IF EXISTS ${catalogo}.silver.credit_record_transformed;
```

```sql
%python
# REMOVE DATA (Silver)
dbutils.fs.rm(f"{ruta}/tablas/application_record_transformed", True)
dbutils.fs.rm(f"{ruta}/tablas/credit_record_transformedS", True)
```
