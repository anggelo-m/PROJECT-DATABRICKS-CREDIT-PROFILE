##Grants

```sql
-- Grants en Catalog y Schemas (GRANT es idempotente, no falla si ya existe)
GRANT ALL PRIVILEGES ON CATALOG credit_catalog TO Data_Engineers;
GRANT ALL PRIVILEGES ON CATALOG credit_catalog TO Data_Scientis;
GRANT SELECT ON SCHEMA credit_catalog.silver TO Data_Scientis;
GRANT SELECT ON SCHEMA credit_catalog.golden TO Data_Scientis;
GRANT CREATE ON SCHEMA credit_catalog.silver TO Data_Engineers;
GRANT CREATE ON SCHEMA credit_catalog.golden TO Data_Engineers;
```

##External Locations

```sql
-- Grants en External Locations (GRANT es idempotente)
GRANT READ FILES
ON EXTERNAL LOCATION `exlt-raw`
TO Data_Engineers;

GRANT READ FILES
ON EXTERNAL LOCATION `exlt-bronze`
TO Data_Engineers;

GRANT READ FILES
ON EXTERNAL LOCATION `exlt-silver`
TO Data_Engineers;

GRANT READ FILES
ON EXTERNAL LOCATION `exlt-golden`
TO Data_Engineers;

GRANT READ FILES
ON EXTERNAL LOCATION `exlt-silver`
TO Data_Scientis;

GRANT READ FILES
ON EXTERNAL LOCATION `exlt-golden`
TO Data_Scientis;
```
